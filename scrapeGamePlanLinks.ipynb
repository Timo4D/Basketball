```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import io

driver = webdriver.Chrome()

# driver.get("https://www.basketball-bund.net/static/#/liga/39952/statistik")
driver.get("https://www.basketball-bund.net/index.jsp?Action=103&liga_id=39952&defaultview=1")
driver.implicitly_wait(5)

accept = driver.find_element(by=By.LINK_TEXT, value="AKZEPTIEREN")
accept.click()

links = []

for i in range(1, 8):
    print(i)

    table = driver.find_element(by=By.XPATH, value="/html/body/div/div[2]/div[5]/div/form/table[3]")

    for row in table.find_elements(By.TAG_NAME, "tr"):
        try:
            link = row.find_elements(By.TAG_NAME, "a")[0].get_attribute("href")
            if "spielplan" in link:
                        print("Guter Link")
                        links.append(link)
            else :
                print("Müll")
        except:
            print("No link")

            
    print("next")
    if i == 1:
             next_button = driver.find_element(by=By.XPATH, value="/html/body/div/div[2]/div[5]/div/form/table[4]/tbody/tr/td/table/tbody/tr[2]/td/table/tbody/tr/td[2]/a[1]")
    else :
           next_button = driver.find_element(by=By.XPATH, value="/html/body/div/div[2]/div[5]/div/form/table[4]/tbody/tr/td/table/tbody/tr[2]/td/table/tbody/tr/td[2]/a[3]")
    next_button.click()


print("Links:")
for link in links:
    print(link)




driver.quit()
```
