```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By

links = pd.read_csv('links.csv', header=0).to_numpy()
```

```python
from io import StringIO


spiele = pd.DataFrame(columns=['Nummer', 'Spieltag', 'Datum', 'Heim', 'Gast', 'Ergebnis', '1. Viertel', 'Halbzeit', '3. Viertel', 'OT', 'Heimspieler', 'Gastspieler'])
spieler = pd.DataFrame(columns=['Team', 'Spiel','Nachname', 'Vorname', 'Punkte', 'FWVersuche', 'FWTreffer', '2er', '3er', 'Fouls'])


link = links[0][0]
i = 0

for row in links:
    link = row[0]
    print(link)
    i += 1
    print(i)
    driver = webdriver.Chrome()

    driver.get(link)
    driver.implicitly_wait(5)

    spiel_nummer = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[1]').text
    print(spiel_nummer)
    spiel_spieltag = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[2]').text
    print(spiel_spieltag)
    spiel_datum = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[3]').text
    print(spiel_datum)
    spiel_heim = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[4]').text
    print(spiel_heim)
    spiel_gast = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[5]').text
    print(spiel_gast)
    spiel_ergebnis = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[6]').text
    print(spiel_ergebnis)
    spiel_1_viertel = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[7]').text
    print(spiel_1_viertel)
    spiel_halbzeit = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[8]').text
    print(spiel_halbzeit)
    spiel_3_viertel = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[9]').text
    print(spiel_3_viertel)
    spiel_ot = driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[10]').text
    print(spiel_ot)


    print('Heimspieler')
    spieler_heim = driver.find_element(By.XPATH, '/html/body/form[2]/table[2]')
    spiel_heimspieler = pd.read_html(StringIO(spieler_heim.get_attribute('outerHTML')), header=[0])[0] 
    spiel_heimspieler = spiel_heimspieler.drop(spiel_heimspieler.index[0:8])
    spiel_heimspieler['Team'] = spiel_heim
    spiel_heimspieler['Spiel'] = spiel_nummer
    print(len(spiel_heimspieler))


    print('Gastspieler')
    spieler_gast = driver.find_element(By.XPATH, '/html/body/form[3]/table[2]')
    spiel_gastspieler = pd.read_html(StringIO(spieler_gast.get_attribute('outerHTML')), header=[0])[0]
    spiel_gastspieler = spiel_gastspieler.drop(spiel_gastspieler.index[0:8])
    spiel_gastspieler['Team'] = spiel_gast
    spiel_gastspieler['Spiel'] = spiel_nummer
    print(len(spiel_gastspieler))

    spiel = pd.DataFrame({'Nummer': [spiel_nummer], 'Spieltag': [spiel_spieltag], 'Datum': [spiel_datum], 'Heim': [spiel_heim], 'Gast': [spiel_gast], 'Ergebnis': [spiel_ergebnis], '1. Viertel': [spiel_1_viertel], 'Halbzeit': [spiel_halbzeit], '3. Viertel': [spiel_3_viertel], 'OT': [spiel_ot], 'Heimspieler': [len(spiel_heimspieler)], 'Gastspieler': [len(spiel_gastspieler)]})


    spiele = pd.concat([spiele, spiel], ignore_index=True)
    spiel_spieler = pd.concat([spiel_heimspieler, spiel_gastspieler], ignore_index=True)
    spieler = pd.concat([spieler, spiel_spieler], ignore_index=True)



    driver.quit()
spiele.head()
```
