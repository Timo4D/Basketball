```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By

links = pd.read_csv('links.csv', header=0).to_numpy()
```

```python
from io import StringIO

driver = webdriver.Chrome()

link = links[0][0]

# for row in links:
# link = row[0]
print(link)

driver.get(link)
driver.implicitly_wait(5)

print('Nummer: ', driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[1]').text)
print('Spieltag: ', driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[2]').text)
print('Datum: ', driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[3]').text)
print('Heim: ', driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[4]').text)
print('Gast: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[5]').text)
print('Ergebnis: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[6]').text)
print('1. Viertel: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[7]').text)
print('Halbzeit: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[8]').text)
print('3. Viertel: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[9]').text)
print('OT: ',driver.find_element(By.XPATH, '/html/body/form[1]/table[2]/tbody/tr[2]/td[10]').text)


print('Heimspieler')
spieler_heim = driver.find_element(By.XPATH, '/html/body/form[2]/table[2]')
heimspieler = pd.read_html(StringIO(spieler_heim.get_attribute('outerHTML')), header=[0])[0] 
heimspieler = heimspieler.drop(heimspieler.index[0:8])
print(len(heimspieler))

print('Gastspieler')
spieler_gast = driver.find_element(By.XPATH, '/html/body/form[3]/table[2]')
gastspieler = pd.read_html(StringIO(spieler_gast.get_attribute('outerHTML')), header=[0])[0]
gastspieler = gastspieler.drop(gastspieler.index[0:8])
print(len(gastspieler))


driver.quit()
```
